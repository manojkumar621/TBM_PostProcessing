```python
import os
import shutil

# Define the directories
dir_A = '/blue/pinaki.sarder/manojkumargalla/PostProcess/data/WSI_tiles_new/temp/'
dir_B = '/blue/pinaki.sarder/manojkumargalla/PostProcess/data/Filtered_image_tiles_new/temp'
dir_C = '/blue/pinaki.sarder/manojkumargalla/PostProcess/data/retrain/training_wsi_tiles_original'
dir_D = '/blue/pinaki.sarder/manojkumargalla/PostProcess/data/retrain/training_wsi_tiles_filtered'

# Create directories C and D if they don't exist
os.makedirs(dir_C, exist_ok=True)
os.makedirs(dir_D, exist_ok=True)

# Get the list of filenames in directory B and directory A
files_in_B = set(os.listdir(dir_B))
files_in_A = set(os.listdir(dir_A))

# Copy files from A to C that also exist in B
for filename in files_in_A:
    if filename in files_in_B:
        shutil.copy(os.path.join(dir_A, filename), os.path.join(dir_C, filename))

# Copy files from B to D that also exist in A
for filename in files_in_B:
    if filename in files_in_A:
        shutil.copy(os.path.join(dir_B, filename), os.path.join(dir_D, filename))

print(f"Files from {dir_A} matching filenames in {dir_B} have been copied to {dir_C}.")
print(f"Files from {dir_B} matching filenames in {dir_A} have been copied to {dir_D}.")
```

```python
import os
from skimage.io import imread
import skimage.color as color

def count_image_types(folder_path):
    grayscale_count = 0
    rgb_count = 0
    
    # Iterate over each file in the folder
    for filename in os.listdir(folder_path):
        if filename.endswith(".jpg") or filename.endswith(".png"):
            file_path = os.path.join(folder_path, filename)
            image = imread(file_path)
            
            # Check if the image is grayscale or RGB
            if len(image.shape) == 2:  # Grayscale images have 2 dimensions
                grayscale_count += 1
            elif len(image.shape) == 3:  # RGB images have 3 dimensions
                if image.shape[2] == 3:  # Ensure it's an RGB image (not RGBA)
                    rgb_count += 1
    
    return grayscale_count, rgb_count

# Example usage
folder_path = "/home/manojkumargalla/ResUNet/data/train/input_crop/"
grayscale_count, rgb_count = count_image_types(folder_path)
print(f"Number of grayscale images: {grayscale_count}")
print(f"Number of RGB images: {rgb_count}")
```

```python
import os
from PIL import Image

def convert_tif_to_png(directory):
    # Create an output directory for PNG files
    output_dir = os.path.join(directory, "converted_pngs")
    os.makedirs(output_dir, exist_ok=True)
    files = os.listdir(directory)
    print(files)
    print(len(files))
    # Loop through all files in the directory
    for filename in os.listdir(directory):
        if filename.endswith(".tif") and not filename.startswith("._"):
            print(filename)
            # Full path of the .tif file
            tif_path = os.path.join(directory, filename)
            
            # Open the image file
            with Image.open(tif_path) as img:
                # Convert image to RGB if necessary
                if img.mode in ("RGBA", "P"):
                    img = img.convert("RGB")
                
                # Create the output .png filename
                png_filename = os.path.splitext(filename)[0] + ".jpg"
                png_path = os.path.join(output_dir, png_filename)
                
                # Save the image as .png
                img.save(png_path, "JPEG")
            
            print(f"Converted {filename} to {png_filename}")

# Specify the directory containing .tif files
directory = "/home/manojkumargalla/ResUNet/data/S-2106-003588_PAS_1of2_png_files"
convert_tif_to_png(directory)
```

```python
import os
import shutil

def move_png_files(source_dir, destination_dir):
    # Create the destination directory if it doesn't exist
    os.makedirs(destination_dir, exist_ok=True)
    
    # Loop through all files in the source directory
    for filename in os.listdir(source_dir):
        if filename.endswith(".tif"):
            # Full path of the source file
            source_file = os.path.join(source_dir, filename)
            
            # Full path of the destination file
            destination_file = os.path.join(destination_dir, filename)
            
            # Move the file
            shutil.move(source_file, destination_file)
            print(f"Moved: {filename} to {destination_dir}")

# Example usage:
source_directory = "/home/manojkumargalla/ResUNet/data/nikhil_annos"
destination_directory = "/home/manojkumargalla/ResUNet/data/nikhil_annos/pngs"

move_png_files(source_directory, destination_directory)
```

```python
import os
from PIL import Image

def convert_rgba_to_rgb_in_place(directory):
    # Loop through all files in the directory
    for filename in os.listdir(directory):
        if filename.lower().endswith((".png", ".tif", ".jpg", ".jpeg")):  # Add other image formats if needed
            # Full path of the source file
            image_path = os.path.join(directory, filename)
            
            # Open the image file
            with Image.open(image_path) as img:
                # Check if the image is in RGBA format
                if img.mode == 'RGBA':
                    # Convert the image to RGB
                    img = img.convert('RGB')
                    
                    # Save the converted image, replacing the original file
                    img.save(image_path)
                    
                    print(f"Converted {filename} to RGB format and replaced the original file")
                else:
                    print(f"{filename} is already in RGB format")

# Example usage:
directory = "/home/manojkumargalla/ResUNet/data/train_100/input_crop"
convert_rgba_to_rgb_in_place(directory)
```
